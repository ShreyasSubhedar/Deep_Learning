#Classes in Python


```
class MobilePhones:
  """ This class is maily used for Mobile Phones """
  def __init__(self,name,is_android,screen_size, rating=-1):
    self.name= name
    self.is_android=is_android
    self.screen_size = screen_size
    self.rating = rating
  def checkRating(self):
    return self.rating > 30

new_phone = MobilePhones("MOto G5",True,6.3)
print(new_phone.is_android,new_phone.name,new_phone.screen_size,new_phone.rating)
new_phone.rating = 28
print(new_phone.is_android,new_phone.name,new_phone.screen_size,new_phone.rating)
print(new_phone.checkRating())
print(MobilePhones.__doc__)
#iphone is class which inheriate MobilePhones
class iPhone(MobilePhones):
  def __init__(self,name):
    MobilePhones.__init__(self,name,False,4.3)

new_iphone= iPhone('Iphone5s')
print(new_iphone.name,new_iphone.rating, new_iphone.screen_size,new_iphone.is_android)
```

```
X = np.array([[1,2,3,4,5],[43,54,6,2,64],[34,4352,23,346,45]])
print(X)

for i in range(X.shape[0]):
  for j in range(X.shape[1]):
    print(X[i][j],end=" ")
  print()


print(X[0,:])
print(X[0:2,:])


#zeros to get zero matrix

arr = np.zeros([5,5])
print(arr)


# eye to get identity matrix

arr1 = np.eye(5,5)
print(arr1)

#random to get matrix with random number

arr2 = np.random.random([5,5])
print(arr2)

#T for transpose
z= X.T
print(z)


#reshape the matrix

z=z.reshape(1,15)
print(z)


# arrage = range  for nd array

A= np.arange(5)
B = np.arange(5)
print(A+B)
print(A-B)
print(A*B)
#print(A/B)

print(A,A+1)  # scalar value is added to vector the broadcasting concept

A = np.random.random((2,3))
print(A)
print()
print(A+1)
A = np.round(np.random.random((2,3))*10)
print(A)

# inner product is multiplying the respe. position number and then add 

A = [1,2,3]
B= [-1,0,1]
print(np.inner(A,B))

# outer product is multiply an element to complete array 

print(np.outer(A,B))

# dot product 
A = np.ones((2,3))
B = np.ones((3,2))
print(np.dot(A,B))
print(np.dot(A.T,B.T))

print(A.sum(axis=1))
```
