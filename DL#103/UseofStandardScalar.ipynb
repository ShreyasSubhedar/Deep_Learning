```
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

import numpy as np
```

```
R= np.random.random([100,1])
```

```
plt.plot(R)
plt.show()
```

INItialize the scalar instance

```
scalar = StandardScaler()
```

```
scalar.fit(R)
```

```
RT= scalar.transform(R)
```

Finding mean of standardized R i.e RT



```
np.mean(RT)
```

Finding std deviation of standardized R i.e RT



```
np.std(RT)
```

```
plt.plot(RT)
plt.show()
```

# Conclusion:
plots of RT and R looks identical <br>
only range is differnt<br>
Standardisation change the mean = 0 and std deviation = 1
