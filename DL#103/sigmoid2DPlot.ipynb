# Plotting Sigmoid Neuron

```
import numpy as np
import matplotlib.pyplot as plt
```

```
def sigmoid(w,x,b):
  return 1/(1+np.exp(-(w*x+b)))
```

```

w=-0.1    #@param {type: "slider", min: -2, max: 2, step:0.1}
b=5   #@param {type: "slider", min: -2, max: 5, step:0.1}
X=np.linspace(-10,10,100) 
Y = sigmoid(w,X,b)
plt.plot(X,Y)
plt.show()
```

#Conclution:
Bias changes the shift of the sigmoid function
