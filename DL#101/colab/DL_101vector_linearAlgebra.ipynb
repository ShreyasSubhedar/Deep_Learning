#Vectors in Python

```
import matplotlib.pyplot as plt
import random
def plotVector(vect):
  color=['r','g','b','y']
  i=0
  m =0
  for v in vect:
    plt.quiver(v[0],v[1],v[2],v[3],scale_units="xy",angles="xy",scale=1,color=color[i%len(color)])
    m = max(v[0],v[1],v[2],v[3],m)
    i = i+1
  m+=5
  plt.xlim(-m,m)
  plt.ylim(-m,m)
  plt.show()
```

```
import numpy as np

import matplotlib.pyplot as plt

plt.quiver(0,0,3,4)
plt.show()
```

```
plt.quiver(0,0,3,4,scale_units="xy",angles="xy",scale=1)
plt.show()
```

```
plt.quiver(0,0,3,4,scale_units="xy",angles="xy",scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```
plt.quiver(0,0,3,4,scale_units="xy",angles="xy",scale=1,color='r')
plt.quiver(0,0,-3,4,scale_units="xy",angles="xy",scale=1,color='g')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```
plotVector([(0,5,12,8),(0,0,-3,4),(2,2,5,5),(0,0,2,11)])
```

#Linear Algebra : Vector addition and Substraction

```
vector = [(0,0,3,4),(0,0,-3,4),(0,0,-3,-2),(0,0,4,-1)] # as these are all tuples we cannot perform any operation easily on'em
plotVector(vector)
```

```
plotVector([vector[0],vector[3]])
```

```
#converting all into np.array
vector = [np.asarray([0,0,3,4]),np.asarray([0,0,-3,4]),np.asarray([0,0,-3,-2]),np.asarray([0,0,4,-1])]
```

```
plotVector([vector[0],vector[3]])
```

```
plotVector([vector[0],vector[0]+vector[3],vector[3]])
```

```
plotVector([vector[0],vector[0]+vector[2],vector[2]])
```

```
plotVector([vector[1],vector[1]+vector[2],vector[2]])
```
